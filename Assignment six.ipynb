**Task 2**

Write a python function that takes the limit as an argument of the Fibonacci series and prints till
that limit.

Function Call:

fibonacci(10)

Output:

0 1 1 2 3 5 8

Function Call:

fibonacci(5)

Output:

0 1 1 2 3 5

````python
#2
def fibonacci(limit):
    lis_1 = [0,1]
    while True:
        
        term = lis_1[-1] + lis_1[-2]
        
        if term > limit:
            break
        
        else:
            lis_1.append(term)
            
    for i in lis_1:
        print(i, end=" ")
fibonacci(5)
````
